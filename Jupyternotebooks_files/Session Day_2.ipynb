```python
# weight in kg/height in m and its square
```

```python
weight = float(input("What is your weight in Kg?"))
height = float(input("What is your height?"))
```

```python
name = input("What is your name?")
```

```python
bmi = weight/height**2
bmi
```

```python
print(name, "your BMI is: ",bmi)
```
