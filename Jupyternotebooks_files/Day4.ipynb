```python
kashti = sns.load_dataset('titanic')
kashti
```

```python
sns.barplot(x="who",y="alone", hue= "sex", data=kashti)
plt.title("Gender ka Plot")
# sns.set_style("dark")
plt.show()
```

```python
sns.barplot(x="sex",y="alone", hue= "who", data=kashti, order=["female","male"])
plt.title("Gender ka Plot")
# sns.set_style("dark")
plt.show()
```

```python
sns.barplot(x="sex",y="alone", hue= "who", data=kashti, order=["female","male"], color ='grey')
plt.title("Gender ka Plot")
# sns.set_style("dark")
plt.show()
```

```python
sns.barplot(x="sex",y="alone", hue= "who", data=kashti, order=["female","male"], color ='grey', ci=None)
plt.title("Gender ka Plot")
# sns.set_style("dark")
plt.show()
```

```python
sns.barplot(x="sex",y="alone", hue= "who", data=kashti, order=["female","male"], color ='grey', ci=None, palette='pastel')
plt.title("Gender ka Plot")
# sns.set_style("dark")
plt.show()
```

```python
from numpy import median

sns.barplot(x="sex",y="alone", hue= "who", data=kashti, order=["female","male"], color ='grey', estimator=median,palette='pastel')
plt.title("Gender ka Plot")
# sns.set_style("dark")
plt.show()
```

```python
from numpy import median

sns.barplot(x="sex",y="alone", hue= "who", data=kashti, order=["female","male"], color ='grey', estimator=median,saturation=0.5)
plt.title("Gender ka Plot")
# sns.set_style("dark")
plt.show()
```

```python
#horizontal plot
import seaborn as sns
from numpy import mean
import matplotlib.pyplot as plt


kashti = sns.load_dataset('titanic')

sns.barplot(x="fare",y="class", hue= "sex", data=kashti, estimator=mean, saturation=1)
plt.title("Gender ka Plot")
# sns.set_style("dark")
plt.show()
```

```python
import seaborn as sns
from numpy import mean
import matplotlib.pyplot as plt


kashti = sns.load_dataset('titanic')

sns.barplot(x="class",y="fare", hue= "sex", data=kashti,
            linewidth=3, facecolor=(0.3, 0.3, 0.3, 0.5),
            errcolor="0.5", edgecolor="0")
plt.title("Gender ka Plot")
# sns.set_style("dark")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np

kashti = sns.load_dataset('titanic')
kashti.head()
```

```python
sns.boxplot(x="survived", y="age",data=kashti)
```

```python
p1= sns.boxplot(x="survived", y="age",showmeans=True,data=kashti)
p1
```

```python
import matplotlib.pyplot as plt


p2 = sns.boxplot(x="survived", 
            y="age",showmeans=True,
            meanprops={"marker":"*",
                       "markersize":"12",
                       "markeredgecolor":"red"},
            data= kashti)



p2.set_title('Box plot of kitnay doobay or kitnay bach gaye',size=16,weight='bold')
p2.set_xlabel('How many survived',size=14,weight='bold')
p2.set_ylabel('Age (years)',size=10,weight='bold')
```

## line plot with multifacets

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


nuqta = sns.load_dataset("dots")
nuqta.head()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


#defining a color pelette
p = sns.color_palette('rocket_r')

#plot lineplot

sns.relplot(data= nuqta, x="time", y="firing_rate",
            hue ="coherence", size="choice", col="align",
            kind="line", size_order=["T1", "T2"], palette=p,
            height=5, aspect=.75, facet_kws=dict(sharex=False))
```
