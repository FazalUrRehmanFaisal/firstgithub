# 2-D arrays

```python
a = np.array([[1,2,3,4,5],[5,4,3,2,1]])
a
```

```python
b = np.array([[6,7],[8,9]])
b
```

```python
c = np.concatenate((a,b), axis=1)
c
```

```python
d = np.array([[1,2],[3,4]])
d
```

```python
f = np.concatenate((b,d), axis=1)
f
```

```python
# to find the number of dimesions
f.ndim
```
